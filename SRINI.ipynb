```python
import yaml
from yaml.loader import SafeLoader
import pandas as pd
import os
import time
import requests
import glob
from pathlib import Path

def configuration():
    with open('config.yaml') as f:
        data = yaml.load(f, Loader = SafeLoader)
        interval = data['data_collection_interval']
        isize = data['data_directory_isize']
        datadir = data['data_dir']
        retention = data['data_retention']
        return interval, isize, datadir, retention
def get_data(context):
    response = requests.get('http://localhost:61208/api/3/'+context)
    return process_response(response)


def process_response(response):
    data = response.json()
    df = pd.DataFrame.from_dict(data, orient='index')
    return df

def writer(df, context, datadir):
    df.to_csv(datadir + context + '.csv', mode='a', header=False)
    return
```

```python

def retention_check(datadir, retention = 2):
    files = glob.glob('psaz/psaz_data.*')
    files.sort(key=lambda x: int(x.split('.')[-1]))
    high = int(files[-1].split('.')[-1])
    low = int(files[0].split('.')[-1])
    #remove folders which are older than retention starting from old
    #if high - low > retention:
    for i in range(low, high - retention):
        os.remove(datadir + 'psaz_data.' + str(i))
     
    return high,low
```

```python
def depth(d):
    if hasattr(d, "values"):
        return 1 + max(map(depth, d.values()))
    if hasattr(d, "__len__") and not hasattr(d, "lower"):
        return 1 + max(map(depth, d))
    return 0
```

```python
def granularize(data, granularity):
    if "".join(filter(lambda x: not x.isdigit(), granularity)) == 'h':
        multiplier = 3600
    elif "".join(filter(lambda x: not x.isdigit(), granularity)) == 'm':
        multiplier = 60
    elif "".join(filter(lambda x: not x.isdigit(), granularity)) == 'd':
        multiplier = 86400
    else:
        multiplier = 1 #considering to be seconds
    
    # find the number in granularity
    number = int("".join(filter(lambda x: x.isdigit(), granularity)))
    
    interval = int(number * multiplier)
    output = pd.DataFrame()
    high = int(float(data.iloc[-1]['time']))
    print(interval)
    low = int(float(high - interval))
    print(low,high)
    for i in range(0,len(data)):
        
        tmp = data.query('time > @low and time < @high')
        if tmp.empty: break
        tmp = tmp.mean()
        output = output.append(tmp,ignore_index=True)
        high -= interval
        low -= interval
    return output
    
```
